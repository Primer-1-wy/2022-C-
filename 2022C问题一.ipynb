```python
from tqdm import tqdm
import re
tqdm.pandas()

def clear_id(s):
    ss = str(s)
    n = int(re.findall(r"\d+",ss)[0])
    return n
file2['id'] = file2['文物采样点'].progress_apply(clear_id)
file1['id'] = file1['文物编号'].progress_apply(clear_id)
```

```python
file2.fillna(value=0,inplace=True)
list1=list(file2.columns)
list1.remove("文物采样点")
list1.remove("id")

file2['sum'] = file2.iloc[:,1:15].sum(axis=1)

file2
```

```python
file_data = pd.merge(file1,file2,on='id')
weathering_data = file_data[file_data['表面风化']=='风化']
no_weathering_data = file_data[file_data['表面风化']=='无风化']
```

```python
Gao_data = weathering_data[weathering_data['类型']=='高钾']
Qian_data = weathering_data[weathering_data['类型']=='铅钡']
no_Gao_data = no_weathering_data[no_weathering_data['类型']=='高钾']
no_Qian_data = no_weathering_data[no_weathering_data['类型']=='铅钡']
```

```python
Gao_data.describe()
```

```python
Qian_data.describe()
```

```python
mean_df = pd.DataFrame(columns=['风化高钾的元素均值','无风化高钾的元素均值','风化铅钡的元素均值','无风化铅钡的元素均值'])
mean_df['风化高钾的元素均值'] = list(Gao_data.describe().iloc[1])[2:]
mean_df['无风化高钾的元素均值'] = list(no_Gao_data.describe().iloc[1])[2:]
mean_df['风化铅钡的元素均值'] = list(Qian_data.describe().iloc[1])[2:]
mean_df['无风化铅钡的元素均值'] = list(no_Qian_data.describe().iloc[1])[2:]
```

```python
mean_df.index = list(Gao_data.describe().columns)[2:]
```

```python
# 0值填充缺失值
mean_df = mean_df.fillna(0)
mean_df['高钾风化前后差值'] = mean_df['风化高钾的元素均值']-mean_df['无风化高钾的元素均值']
mean_df['铅钡风化前后差值'] = mean_df['风化铅钡的元素均值']-mean_df['无风化铅钡的元素均值']
mean_df
```

```python
mean_df['高钾风化前后变动率']=mean_df['高钾风化前后差值']/mean_df['风化高钾的元素均值']
mean_df['铅钡风化前后变动率']=mean_df['铅钡风化前后差值']/mean_df['风化铅钡的元素均值']
mean_df
```

```python
mean_df.to_csv("E:/2022国赛/C题/问题一代码及图片/问题一代码及图片/data/元素变化情况.csv")
mean_df=mean_df.drop('sum')
```

```python
#柱状图
import numpy as np
import matplotlib.pyplot as plt
f_list = mean_df['风化高钾的元素均值'][0:]
s_list = mean_df['无风化高钾的元素均值'][0:]
n = len(f_list)
x = np.arange(n)


total_width= 5    # 有多少个类型，只需更改n即可，比如这里我们对比了四个，那么就把n设成4
width = total_width / n
# x = x - (total_width - width) / 2


plt.figure(figsize=(15, 7),dpi=300)
plt.bar(x-width/2, f_list, color = "r",width=width,label='风化')
plt.bar(x + width/2, s_list, color = "g",width=width,label='无风化')
plt.xlabel("化学成分")
plt.ylabel("数值")
plt.legend(loc = "best")
plt.xticks(range(n),mean_df.index[0:],rotation=30)
plt.savefig(picpath+'/高钾风化前后元素对比柱状图.png')
plt.show()
```
